Load in Weather Data

```python
# We'll do precipitation files first
```

```python
precip_files = [file for file in os.listdir(os.getcwd() + '\\Weather Data') if file.startswith('pr')]
```

```python
precip_df = []

for file in precip_files:
    precip_df.append(pd.read_csv('.\\Weather Data\\' + file, skiprows = 1))
```

```python
precip_df_clean = [df[df.columns[1]] for df in precip_df]
```

```python
precip_df_concat = pd.concat(precip_df_clean, axis = 1)
```

```python
precip_df_concat['Year'] = np.arange(1901, 2022)

precip_df_concat.head()
```

```python
# Now for mean temperature files
# I'm going to be copy and pasting here but using Hadley Wickhams (Chief Scientist at R Studio) rules of thumb 
# for when to write a function
# I won't write a function seen as I'll be copying and pasting just once
# Hadly Wickham recommends writing a function when you copy and paste more than twice
```

```python
temper_files = [file for file in os.listdir(os.getcwd() + '\\Weather Data') if file.startswith('tas')]
```

```python
temper_df = []

for file in temper_files:
    temper_df.append(pd.read_csv('.\\Weather Data\\' + file, skiprows = 1))
```

```python
temper_df_clean = [df[df.columns[1]] for df in temper_df]
```

```python
temper_df_concat = pd.concat(temper_df_clean, axis = 1)
```

```python
temper_df_concat['Year'] = np.arange(1901, 2022)

temper_df_concat.head()
```

```python
# Write out the files

precip_df_concat.to_csv('precipitation.csv')
temper_df_concat.to_csv('temperature.csv')
```
